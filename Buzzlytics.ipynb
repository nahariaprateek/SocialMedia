```python
import numpy as np
from google.colab import autoviz

def scatter_plots(df, colname_pairs, figscale=1, alpha=.8):
  from matplotlib import pyplot as plt
  plt.figure(figsize=(len(colname_pairs) * 10 * figscale, 5 * figscale))
  for plot_i, (x_colname, y_colname) in enumerate(colname_pairs, start=1):
    ax = plt.subplot(1, len(colname_pairs), plot_i)
    df.plot(kind='scatter', x=x_colname, y=y_colname, s=(10 * figscale), alpha=alpha, ax=ax)
    ax.spines[['top', 'right',]].set_visible(False)
  plt.tight_layout()
  return autoviz.MplChart.from_current_mpl_state()

chart = scatter_plots(tags_df, *[[['Views', 'Likes'], ['Likes', 'Comments'], ['Comments', 'Shares'], ['Shares', 'Usage']]], **{})
chart
```

```python
def get_duration_category(duration):
    if duration <= 30:
        return "Less than 30s"
    elif duration <= 60:
        return "31s to 60s"
    elif duration <= 300:
        return "61s to 300s"
    else:
        return "More than 300s"
```

```python
videos_df['Duration_Category'] = videos_df['Duration'].apply(get_duration_category)
```

```python
grouped_data = videos_df.groupby([videos_df['Created'].dt.year, 'Duration_Category']).size().reset_index(name='Number of Videos')
grouped_data
```

```python
grouped_data_title = videos_df.groupby([videos_df['Created'].dt.year, 'Duration_Category']).agg({
    'Description': list
}).reset_index()
grouped_data_title
```

```python
def categorize_duration(duration):
    if duration < 30:
        return 'Less than 30s'
    elif 30 <= duration <= 60:
        return '31s to 60s'
    elif 61 <= duration <= 300:
        return '61s to 300s'
    else:
        return 'More than 300s'

videos_df['Duration_Category'] = videos_df['Duration'].apply(categorize_duration)
agg_data = videos_df.groupby('Duration_Category').agg({
    'Views': 'sum',
    'Comments': 'sum',
    'Likes': 'sum',
    'Shares': 'sum'
}).reset_index()
agg_data = agg_data.sort_values(by='Views', ascending=False)
most_popular_category = agg_data.iloc[0]['Duration_Category']
print(f"The most popular category based on views is: {most_popular_category}")
print(agg_data)
```

```python
agg_data #for most popular category
```

```python
if 'Duration_Category' in agg_data.columns:
    agg_data.set_index('Duration_Category', inplace=True)
```

```python
# Sorting data for better visualization
agg_data = agg_data.sort_values(by='Views')

# Colors for the different metrics
colors = ['blue', 'green', 'red', 'orange']

# Creating a 2x2 grid of bar charts
fig, ax = plt.subplots(2, 2, figsize=(15, 10))

metrics = ['Views', 'Comments', 'Likes', 'Shares']
titles = ['Total Views by Duration Category', 'Total Comments by Duration Category', 'Total Likes by Duration Category', 'Total Shares by Duration Category']

for i, metric in enumerate(metrics):
    row = i // 2
    col = i % 2
    agg_data[metric].plot(kind='bar', ax=ax[row, col], color=colors[i], stacked=True)
    ax[row, col].set_title(titles[i])
    ax[row, col].set_ylabel(metric)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
